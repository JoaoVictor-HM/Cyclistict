```python
import pandas as pd

data_2020 = pd.read_csv('2020_data.csv', low_memory=False)
data_2021 = pd.read_csv('2021_data.csv', low_memory=False)
data_2022 = pd.read_csv('2022_data.csv', low_memory=False)
```

### Create a new column for the ride total duration

```python
data_2020['started_at'] = data_2020['started_at'].astype('datetime64[ns]')
data_2020['ended_at'] = data_2020['ended_at'].astype('datetime64[ns]')
```

```python
data_2021['started_at'] = data_2021['started_at'].astype('datetime64[ns]')
data_2021['ended_at'] = data_2021['ended_at'].astype('datetime64[ns]')
data_2022['started_at'] = data_2022['started_at'].astype('datetime64[ns]')
data_2022['ended_at'] = data_2022['ended_at'].astype('datetime64[ns]')
```

```python
data_2020['ride_total_time'] = (data_2020['ended_at']-data_2020['started_at'])
```

```python
data_2021['ride_total_time'] = (data_2021['ended_at']-data_2021['started_at'])
data_2022['ride_total_time'] = (data_2022['ended_at']-data_2022['started_at'])
```

```python
data_bases = [data_2020, data_2021, data_2022]
```

#### Get the duration of the ride in hours

```python
for database in data_bases:
    database['ride_total_time_hours'] = database['ride_total_time'] / pd.Timedelta(hours=1)
```

#### Get the ride weekday

```python
from datetime import datetime
for database in data_bases:
    database['ride_weekday'] = pd.to_datetime(database['started_at']).dt.day_name()
    database['ride_weekday_number'] = database['started_at'].dt.dayofweek
    
    
```

```python
data_2020
```

#### Calculate Distance between End and Start Station

```python
import h3
for database in data_bases:
    database['ride_distance'] = database.apply(lambda row: h3.point_dist((row['start_lat'], row['start_lng']), (row['end_lat'], row['end_lng']), unit='km'), axis=1)
```

```python
ordered_2020 = data_2020.sort_values(by=['ride_weekday_number'])
ordered_2020 
```

### Total Rides per Weekdays by User Type

```python
grouped_2020 = ordered_2020.groupby(['ride_weekday', 'member_casual']).count()['ride_id']
grouped_2020
```

```python
grouped_2021 = data_2021.groupby(['ride_weekday', 'member_casual']).count()['ride_id']
grouped_2021
```

```python
grouped_2022 = data_2022.groupby(['ride_weekday', 'member_casual']).count()['ride_id']
grouped_2022
```

```python
grouped_2020 = ordered_2020.groupby(['member_casual']).mean()['ride_total_time_hours']
grouped_2020 
```

### Get the number of rides on each route (starting station - end station) by user type

```python
grouped_2020 = data_2020.groupby(
    ['member_casual', 'start_station_name', 'end_station_name']).count()['ride_id']
sorted_2020 = grouped_2020.sort_values(ascending=False)
sorted_2020.groupby('member_casual').head(3)
```

```python
grouped_2021 = data_2021.groupby(
    ['member_casual', 'start_station_name', 'end_station_name']).count()['ride_id']
sorted_2021 = grouped_2021.sort_values(ascending=False)
sorted_2021.groupby('member_casual').head(3)
```

```python
grouped_2022 = data_2022.groupby(
    ['member_casual', 'start_station_name', 'end_station_name']).count()['ride_id']
sorted_2022 = grouped_2022.sort_values(ascending=False)
sorted_2022.groupby('member_casual').head(4)
```
